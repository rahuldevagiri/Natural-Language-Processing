# 1. Learning Rate

### Concept

The learning rate controls how large the weight updates are during training.

A small learning rate learns slowly but may be more stable.

A large learning rate learns faster but may overshoot the optimum solution.

### Typical Values

```text
0.1      Very large
0.01     Large
0.001    Common default
0.0001   Small
```

### PyTorch

```python
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)
```

### Keras

```python
optimizer = keras.optimizers.Adam(
    learning_rate=0.001
)
```

### Analogy

Imagine walking down a hill:

- Large steps → faster but easier to trip.
- Small steps → safer but slower.

# 4. Optimiser

### Concept

The algorithm used to update weights.

### Common Choices

| Optimiser | Description |
|------------|------------|
| SGD | Basic gradient descent |
| Adam | Most popular |
| RMSprop | Good for recurrent networks |
| AdamW | Adam with weight decay |

### PyTorch

```python
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)
```

### Keras

```python
optimizer = keras.optimizers.Adam(
    learning_rate=0.001
)
```
### Example

Assume we are building a model that predicts house prices.

**Scenario**:
We give the model input data:
```text
House features → Size, number of rooms, location
True price → 300,000 €
Model prediction → 250,000 €
```
The model is wrong, so we compute an error (loss):
```text
Error = 50,000 €
```
Now the model must adjust its internal parameters (weights) to improve future predictions.

**What is the role of the optimiser (Adam)?**

The optimiser (Adam) is the mechanism that decides:

**How to correct the model’s weights after each error?**

So instead of manually changing weights, we say:
> **Use Adam to automatically update weights in a smart way**

---

In General, think of **training as adjusting a recipe**:

- Model = chef
- Weights = ingredients
- Loss = taste error
- Optimiser = method of correcting the recipe
- Learning rate = how strongly you change the recipe

# 7. Number of Layers

### Concept

The depth of the neural network.

### Example

```text
Input (784-dimensional vector)
   ↓
Hidden Layer 1 (128 units)
   ↓
Hidden Layer 2 (64 units)
   ↓
Output Layer (10 units)
```

### PyTorch

```python
self.fc1 = nn.Linear(784,128)
self.fc2 = nn.Linear(128,64)
self.fc3 = nn.Linear(64,10)
```

### Keras

```python
Sequential([
    Dense(128),
    Dense(64),
    Dense(10)
])
```
### Explaination

- In this architecture, two hidden layers are identified because they are the learnable intermediate layers located between the input representation (784-dimensional vector) and the output layer (10 classes). Specifically, the transformations 784 → 128 and 128 → 64 constitute the hidden representation learning stages, while the final mapping 64 → 10 produces the class logits.
- **Class logits** are the raw output scores of the final neural network layer, where each value represents the model’s unnormalised evidence for a particular class before being converted into probabilities via softmax, for instance.

# 13. Random Seed

### Concept

Controls randomness for reproducible experiments.

### PyTorch

```python
torch.manual_seed(42)
```

### Keras / TensorFlow

```python
tf.random.set_seed(42)
```

### Benefit

```text
Run 1 → similar results
Run 2 → similar results
```
### Explaination

Think of training like rolling dice at the start:

- Without seed → dice rolls are unpredictable each time
- With seed → dice rolls follow the same sequence every run

So:

**seed = 42 → “replay the same experiment”**
> Setting a random seed fixes the starting point of all stochastic processes in the training pipeline, ensuring that model initialisation, data shuffling, and other random operations produce identical results across multiple runs

# 15. RNN / LSTM / GRU Specific Hyper-Parameters

## Hidden Size

Number of memory units.

### PyTorch

```python
nn.LSTM(
    input_size=10,
    hidden_size=64
)
```

### Keras

```python
LSTM(64)
```

### Explaination
- The hidden size (e.g., 64) means the LSTM stores its memory as a 64-number vector at each time step.
- At every step in a sequence, the model keeps two internal memories:
   - a hidden state (short-term output memory)
   - a cell state (long-term memory)
   - Both are vectors of size 64.
- Instead of remembering raw inputs (words or features), the LSTM compresses information into these 64 learned values.
- Each of the 64 values does not have a fixed meaning; they are learned automatically to capture useful patterns in the data.
- As the sequence progresses, the LSTM:
  - adds new information from the current input
  -  forgets irrelevant past information
  -  updates its 64-dimensional memory accordingly

- A larger hidden size means:
    - more memory capacity
    - ability to learn more complex patterns

- A smaller hidden size means:
  - less memory capacity
  - faster but more limited learning
---

## Number of Recurrent Layers

### PyTorch

```python
nn.LSTM(
    input_size=10,
    hidden_size=64,
    num_layers=2
)
```

### Keras

```python
Sequential([
    LSTM(64, return_sequences=True),
    LSTM(64)
])
```

### Explaination

- This model uses a stacked LSTM with two recurrent layers, where each layer has a hidden size of 64 and processes sequences made of 10-dimensional input vectors.

- The first LSTM layer reads the raw sequence and converts it into a sequence of 64-dimensional representations, capturing basic temporal patterns such as short-term dependencies and local relationships between time steps.

- The second LSTM layer takes this transformed sequence as input and further refines it, learning more abstract and higher-level temporal structures across the entire sequence.

- Overall, the architecture can be understood as a stepwise process where the model first learns simple sequential patterns and then builds more complex contextual understanding before producing the final output.

---

## Recurrent Dropout

Applies dropout inside the recurrent connections.

### Keras

```python
LSTM(
    64,
    recurrent_dropout=0.2
)
```

### PyTorch

Not directly available in standard LSTM cells. PyTorch doesn't have a dedicated "recurrent dropout" implementation.
What PyTorch gives you:
- ```nn.LSTM(dropout=p)```: drops between layers, standard (non-variational) dropout only
- ```nn.Dropout```: generic, applies a new mask each forward call

> There's no built-in variational/locked dropout, no hidden-to-hidden dropout, nothing specifically designed for recurrent use.
