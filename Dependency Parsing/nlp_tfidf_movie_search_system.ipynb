# NLP Assignment: TF-IDF Search System

This notebook builds a small **search / recommendation system** that matches a user query with the most relevant text documents using **TF-IDF** and **cosine similarity**.

## What this notebook does
- Creates a dataset with 10 short text documents
- Preprocesses the text
- Converts text into TF-IDF vectors
- Accepts a user query
- Calculates cosine similarity
- Returns the top 3 most relevant results


## 1. Import libraries

We use:
- `pandas` for handling the dataset
- `nltk` for preprocessing
- `sklearn` for TF-IDF and cosine similarity


```python

import pandas as pd
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## 2. Download NLTK resources

Run this once if your system does not already have these resources.


```python

nltk.download('stopwords')
nltk.download('wordnet')
```

## 3. Create a small dataset

Each document has a title and a short description.


```python

documents = [
    {
        "title": "Galaxy Rescue",
        "description": "A brave astronaut travels across space with intelligent robots to save a lost colony."
    },
    {
        "title": "Healthy Kitchen",
        "description": "A practical guide to healthy food, fresh vegetables, and simple home cooking."
    },
    {
        "title": "Robot Future",
        "description": "Scientists build advanced robots that help humans in a futuristic city."
    },
    {
        "title": "Mountain Escape",
        "description": "A group of friends go on an adventurous trip through mountains, rivers, and forests."
    },
    {
        "title": "Deep Sea Mystery",
        "description": "Divers explore the ocean and discover strange machines hidden under water."
    },
    {
        "title": "Space Journey",
        "description": "An exciting space adventure with aliens, spaceships, and heroic missions."
    },
    {
        "title": "Fitness Habits",
        "description": "Tips for exercise, balanced meals, healthy snacks, and an active lifestyle."
    },
    {
        "title": "Cooking for Beginners",
        "description": "Easy recipes, cooking basics, and useful ideas for preparing tasty food."
    },
    {
        "title": "AI Companions",
        "description": "Smart machines and social robots become part of everyday human life."
    },
    {
        "title": "Desert Survival",
        "description": "Travelers struggle to survive in a harsh desert filled with danger and mystery."
    }
]

df = pd.DataFrame(documents)
df
```

## 4. Preprocess the text

The preprocessing steps are:
- Convert text to lowercase
- Remove punctuation
- Remove stopwords
- Lemmatize words


```python

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)

df['clean_description'] = df['description'].apply(preprocess_text)
df[['title', 'clean_description']]
```

## 5. Convert documents into TF-IDF vectors

`TfidfVectorizer` converts the cleaned text into numerical vectors so that we can compare documents and queries mathematically.


```python

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['clean_description'])

tfidf_matrix.shape
```

## 6. Enter a user query

You can change the query below and run the cell again.


```python

user_query = "space adventure with robots"
clean_query = preprocess_text(user_query)

print("Original query:", user_query)
print("Processed query:", clean_query)
```

## 7. Calculate cosine similarity

We transform the query using the same TF-IDF vectorizer and then compare it with all document vectors.


```python

query_vector = vectorizer.transform([clean_query])
similarity_scores = cosine_similarity(query_vector, tfidf_matrix).flatten()

df['similarity_score'] = similarity_scores
df[['title', 'similarity_score']].sort_values(by='similarity_score', ascending=False)
```

## 8. Show the top 3 results

These are the most relevant documents for the given query.


```python

top_results = df.sort_values(by='similarity_score', ascending=False).head(3)

for index, row in top_results.iterrows():
    print("Title:", row['title'])
    print("Description:", row['description'])
    print("Similarity Score:", round(row['similarity_score'], 4))
    print("-" * 50)
```
