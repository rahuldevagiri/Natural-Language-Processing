```python
corpus=[
    "papers are made using trees",
    "collection of those papers are made into books",
    "many people like to read books",
    "book give good knowledge",
    "reading books help improve your speaking skills"
]

query="how to improve my speaking skills"
```

```python
from rank_bm25 import BM25Okapi

tokenizer_corpus=[doc.lower().split() for doc in corpus]
tokenizer_query=query.lower().split()
```

```python
bm25=BM25Okapi(tokenizer_corpus)
scores=bm25.get_scores(tokenizer_query)
```

```python
for i,s in enumerate(scores):
    print(f"Doc {i+1} -> score {s:3f} | {corpus[i]}")
```

```python
ranked = sorted(
    zip(corpus, scores), # Combine documents with their scores
    key=lambda x: x[1], # Sort by score (x[1]) in descending order
    reverse=True # Set reverse to True for descending order
)

for rank, (doc, score) in enumerate(ranked, start=1):
    print(f"{rank}. {doc}")
    print(f"   Score: {score:.4f}\n")
```

```python
top_docs = bm25.get_top_n( 
    tokenizer_query, # The tokenized query
    corpus,
    n=3
)

for doc in top_docs:
    print(doc)
```
